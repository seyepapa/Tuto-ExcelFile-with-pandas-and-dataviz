# Data vizualisation 

### Dans cette partie, nous allons utiliser df_3

L'ensemble de données contient des données annuelles sur les flux de migrants internationaux enregistrées par les pays de destination. Les données présentent à la fois les entrées et les sorties selon le lieu de naissance, la nationalité ou le lieu de résidence précédent / suivant, tant pour les étrangers que pour les nationaux. Pour cette leçon, nous allons nous concentrer sur les données d’immigration canadienne.

```python
df_3.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)
```

```python
df_3.head()
```

### Suppressions des colonnes dont nous n'avons pas besoin pour la visualisation

```python
df_3.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_3.head()
```

```python
df_3['Total'] = df_3.sum(axis=1)
```

À présent, le dataframe comporte une colonne supplémentaire qui présente le nombre total d'immigrants de chaque pays dans l'ensemble de données de 1980 à 2013.

```python
df_3.head()
```

```python
Changement des index de notre dataframe avec la methode set_index
```

```python
df_3.set_index('Country', inplace=True)
```

```python
df_3.head()
```

# Line plot

```python
years = list( range(1980,2014))
```

```python
df_3.loc['Algeria', years].plot(kind='line')
plt.xlabel('Année')
plt.ylabel("Nombre d'immigrants")
plt.show()
```

# Area plot

```python
df_3_sorted = df_3.sort_values(['Total'], ascending=False, axis=0) #trie par ordre décroissante de la colonne total
```

```python
df_3_sorted.head()
```

```python
df_3_sorted.head()[years].transpose().plot(kind='area')#transposition des cinq premiers pays qui on plus d'immigrans
plt.title('Top 5')
plt.legend()
plt.show()
```

```python
df_3_sorted.tail(5)[years].transpose().plot(kind='area')#transposition des cinq derniers pays qui on plus d'immigrans
plt.title('last 5')
plt.legend()
plt.show()
```

# Histogramme

C'est une façon de représenter la fréquence de distribution d'une variable

```python
df_3[2013].plot(kind='hist', color='red')
plt.xlabel("Nombre d'immigrants")
plt.ylabel('Nombre de pays')
plt.show()
```

```python
df_3[2009].plot(kind='hist', color='green')
plt.xlabel("Nombre d'immigrants")
plt.ylabel('Nombre de pays')
plt.show()
```

# Pie chart

```python
df_3_sorted['Total'].head().plot(kind='pie', explode=[0.1, 0.02, 0, 0, 0], shadow=True)#transposition des cinq premiers pays qui on plus d'immigrans
plt.axis('equal')
plt.show()
```
